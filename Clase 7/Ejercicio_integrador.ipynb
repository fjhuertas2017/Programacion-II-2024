# Ejercicio: Sistema de gestión de una tienda de productos electrónicos

Imagina que eres el encargado de desarrollar un pequeño sistema para gestionar las ventas de una tienda de productos electrónicos. El sistema debe permitir registrar productos, calcular el costo total de una compra y aplicar descuentos según las condiciones del cliente. Para ello, debes crear un programa que realice las siguientes tareas utilizando **funciones**, **datos complejos**, **estructuras condicionales** y **bucles**.

## Requisitos del programa:

1. **Registrar productos:**
   - Crea una función que permita registrar los productos en el sistema. Cada producto debe tener un nombre, un precio y una cantidad en inventario. Utiliza un **diccionario** para almacenar esta información.

2. **Simulación de una compra:**
   - Crea una función que permita simular una compra. Esta función debe pedir al usuario que ingrese los productos que desea comprar y la cantidad de cada uno.
   - Verifica si la cantidad solicitada está disponible en el inventario. Si no hay suficiente inventario, muestra un mensaje de error.
   - Almacena la compra en una **lista de diccionarios** donde cada diccionario represente un producto y la cantidad comprada.

3. **Aplicar descuentos:**
   - Crea una función que calcule el total de la compra y aplique un descuento si el cliente es **miembro** de la tienda:
     - Si el cliente no es miembro, no se aplica descuento.
     - Si el cliente es miembro, se aplica un 10% de descuento si el total es mayor a $1000.
   - Utiliza una estructura condicional para aplicar el descuento.

4. **Mostrar el resumen de la compra:**
   - Crea una función que imprima un resumen de la compra, incluyendo el total antes y después del descuento (si corresponde).

## Detalles técnicos:
- Usa **funciones** para dividir las responsabilidades (por ejemplo, una función para registrar productos, otra para calcular el total, otra para aplicar descuentos, etc.).
- Usa **estructuras complejas de datos** como listas y diccionarios para gestionar la información de los productos y las compras.
- Usa **bucles** para permitir que el cliente añada varios productos a la compra.
- Usa **estructuras condicionales** para gestionar el descuento y el control de inventario.

## Puntos a evaluar:
- Uso adecuado de **funciones** para organizar el código.
- Correcto manejo de **datos complejos** como listas y diccionarios.
- Implementación eficiente de **estructuras condicionales** para aplicar descuentos y verificar el inventario.
- Uso de **bucles** para procesar múltiples productos en la compra.
